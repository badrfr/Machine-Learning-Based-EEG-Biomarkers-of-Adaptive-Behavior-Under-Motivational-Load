#### Prediction avec ces features : 'mean', 'std', 'min', 'max', 'median', 'skew'

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
from timeit import default_timer as timer

from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, cross_val_predict, StratifiedKFold
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.pipeline import make_pipeline
from sklearn.calibration import CalibratedClassifierCV
from sklearn.feature_selection import SelectKBest, f_classif, VarianceThreshold
from sklearn.impute import SimpleImputer

# Ajout des imports pour les modèles RNN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GRU, Bidirectional, LSTM, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# === 1. Colonnes de features et liste des fichiers ===
feature_cols = ['EEG1', 'EEG2', 'EEG3', 'EEG4', 'EEG5',
                'EEG6', 'EEG7', 'EEG8', 'Connection', 'AUX1&AUX2', 'EVENTa&b']

fichiers = [
    ('AO_10.csv', 1), ('AO_11.csv', 1), ('AO_12.csv', 1), ('AO_13.csv', 1),
    ('AO_14.csv', 1), ('AO_15.csv', 1), ('AO_16.csv', 1), ('AO_17.csv', 1),
    ('AO_18.csv', 1), ('AO_19.csv', 1), ('AO_2.csv', 1), ('AO_22.csv', 1),
    ('AO_23.csv', 1), ('AO_24.csv', 1), ('AO_3.csv', 1), ('AO_4.csv', 1),
    ('AO_6.csv', 1), ('AO_20.csv', 1), ('AO_21.csv', 1), ('AO_8.csv', 1),
    ('AO_9.csv', 1),
    ('SO_1.csv', 0), ('SO_10.csv', 0), ('SO_11.csv', 0), ('SO_12.csv', 0),
    ('SO_13.csv', 0), ('SO_14.csv', 0), ('SO_16.csv', 0), ('SO_17.csv', 0),
    ('SO_19.csv', 0), ('SO_2.csv', 0), ('SO_20.csv', 0), ('SO_23.csv', 0),
    ('SO_24.csv', 0), ('SO_25.csv', 0), ('SO_3.csv', 0), ('SO_4.csv', 0),
    ('SO_5.csv', 0), ('SO_21.csv', 0), ('SO_22.csv', 0), ('SO_6.csv', 0),
    ('SO_7.csv', 0)
]

# === 2. Fonction pour extraire les statistiques de chaque fichier ===
def extraire_features(df, label=None):
    features = {}
    for col in feature_cols:
        if col in df.columns:
            series = pd.to_numeric(df[col].astype(str).str.replace(',', '.'), errors='coerce')
            if series.notna().any():
                features[f'{col}_mean'] = series.mean()
                features[f'{col}_std'] = series.std()
                features[f'{col}_min'] = series.min()
                features[f'{col}_max'] = series.max()
                features[f'{col}_median'] = series.median()
                features[f'{col}_skew'] = series.skew()
            else:
                features.update({f'{col}_{stat}': 0 for stat in ['mean', 'std', 'min', 'max', 'median', 'skew']})
        else:
            features.update({f'{col}_{stat}': 0 for stat in ['mean', 'std', 'min', 'max', 'median', 'skew']})
    if label is not None:
        features['Label'] = label
    return features

# === 3. Chargement et extraction des features de tous les fichiers ===
train_data = []
for nom_fichier, label in fichiers:
    try:
        df = pd.read_csv(nom_fichier, thousands='.', decimal=',', low_memory=False)
        features = extraire_features(df, label)
        train_data.append(features)
    except Exception as e:
        print(f"Erreur avec {nom_fichier}: {e}")
        train_data.append({f'{col}_{stat}': 0 for col in feature_cols for stat in ['mean', 'std', 'min', 'max', 'median', 'skew']} | {'Label': label})

df_train = pd.DataFrame(train_data)

# Handle missing values
imputer = SimpleImputer(strategy='median')
X = pd.DataFrame(imputer.fit_transform(df_train.drop('Label', axis=1)),
                 columns=df_train.drop('Label', axis=1).columns)
y = df_train['Label']

# === 4. Improved Feature Selection ===
# First remove constant features
constant_filter = VarianceThreshold(threshold=0)
X_filtered = constant_filter.fit_transform(X)
X_filtered = pd.DataFrame(X_filtered, columns=X.columns[constant_filter.get_support()])

# === 5. Préparation des données pour RNN ===
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_filtered)
X_rnn = X_scaled.reshape((X_scaled.shape[0], 1, X_scaled.shape[1]))  # Reshape pour RNN
y_encoded = LabelEncoder().fit_transform(y)

# === 6. Définition des modèles RNN ===
def create_gru_model(input_shape):
    model = Sequential([
        GRU(64, input_shape=input_shape, return_sequences=True, dropout=0.2, recurrent_dropout=0.2),
        GRU(32, dropout=0.2, recurrent_dropout=0.2),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
    return model

def create_bilstm_model(input_shape):
    model = Sequential([
        Bidirectional(LSTM(64, return_sequences=True, dropout=0.2, recurrent_dropout=0.2), input_shape=input_shape),
        Bidirectional(LSTM(32, dropout=0.2, recurrent_dropout=0.2)),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
    return model

# === 7. Modèles ML à comparer ===
models = {
    "Random Forest": RandomForestClassifier(n_estimators=100, max_depth=5, min_samples_split=5, random_state=42),
    "XGBoost": XGBClassifier(n_estimators=100, max_depth=3, learning_rate=0.1, random_state=42, eval_metric='logloss'),
    "MLP": make_pipeline(StandardScaler(), MLPClassifier(hidden_layer_sizes=(50,), alpha=0.01, max_iter=500, random_state=42, early_stopping=True)),
    "SVM": make_pipeline(StandardScaler(), CalibratedClassifierCV(SVC(kernel='rbf', C=1, random_state=42), cv=3)),
    "Logistic Regression": make_pipeline(StandardScaler(), LogisticRegression(C=0.1, max_iter=1000, random_state=42)),
    "KNN": make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors=5, weights='distance'))
}

# === 8. Fonction pour évaluer les modèles RNN ===
def evaluate_rnn_model(model_fn, X, y, name, n_splits=5):
    cv = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    results = {
        "Accuracy": [], "F1-score": [], "Precision": [], "Recall": [],
        "Training Time": [], "Testing Time": []
    }
    
    for train_idx, test_idx in cv.split(X, y):
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]
        
        model = model_fn((X_train.shape[1], X_train.shape[2]))
        early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
        
        # Entraînement
        start_train = timer()
        history = model.fit(X_train, y_train, 
                          validation_data=(X_test, y_test),
                          epochs=50, 
                          batch_size=16, 
                          verbose=0,
                          callbacks=[early_stop])
        train_time = timer() - start_train
        
        # Évaluation
        start_test = timer()
        y_pred = (model.predict(X_test) > 0.5).astype(int).flatten()
        test_time = timer() - start_test
        
        # Stocker les résultats
        results["Accuracy"].append(accuracy_score(y_test, y_pred))
        results["F1-score"].append(f1_score(y_test, y_pred))
        results["Precision"].append(precision_score(y_test, y_pred))
        results["Recall"].append(recall_score(y_test, y_pred))
        results["Training Time"].append(train_time)
        results["Testing Time"].append(test_time)
    
    # Calculer les moyennes
    avg_results = {k: np.mean(v) for k, v in results.items()}
    print(f"\n--- {name} ---")
    for metric, value in avg_results.items():
        print(f"{metric}: {value:.4f}")
    
    return avg_results

# === 9. Évaluation des modèles traditionnels ===
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_results = {}

print("\n=== Évaluation des modèles par validation croisée ===")
for name, model in models.items():
    print(f"\n--- {name} ---")
   
    # Initialisation des listes pour stocker les temps
    train_times = []
    test_times = []
    y_pred_total = np.zeros_like(y)
   
    for train_idx, test_idx in cv.split(X_filtered, y):
        X_train, X_test = X_filtered.iloc[train_idx], X_filtered.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
       
        # Entraînement et mesure du temps
        start_train = timer()
        model.fit(X_train, y_train)
        train_time = timer() - start_train
        train_times.append(train_time)
       
        # Prédiction et mesure du temps
        start_test = timer()
        y_pred = model.predict(X_test)
        test_time = timer() - start_test
        test_times.append(test_time)
       
        y_pred_total[test_idx] = y_pred
   
    # Calcul des métriques
    acc = accuracy_score(y, y_pred_total)
    f1 = f1_score(y, y_pred_total)
    precision = precision_score(y, y_pred_total)
    recall = recall_score(y, y_pred_total)
    cm = confusion_matrix(y, y_pred_total)
   
    # Calcul des temps moyens
    avg_train_time = np.mean(train_times)
    avg_test_time = np.mean(test_times)
   
    cv_results[name] = {
        "Accuracy": acc,
        "F1-score": f1,
        "Precision": precision,
        "Recall": recall,
        "Confusion Matrix": cm,
        "Training Time (s)": avg_train_time,
        "Testing Time (s)": avg_test_time
    }

    print(f"Accuracy       : {acc:.3f}")
    print(f"F1-score       : {f1:.3f}")
    print(f"Precision      : {precision:.3f}")
    print(f"Recall         : {recall:.3f}")
    print(f"Training Time  : {avg_train_time:.4f} sec (moyenne par fold)")
    print(f"Testing Time   : {avg_test_time:.4f} sec (moyenne par fold)")
    print("Confusion Matrix:")
    print(cm)
    print(classification_report(y, y_pred_total, target_names=["Sans objectif", "Avec objectif"]))

# === 10. Évaluation des modèles RNN ===
gru_results = evaluate_rnn_model(create_gru_model, X_rnn, y_encoded, "GRU")
bilstm_results = evaluate_rnn_model(create_bilstm_model, X_rnn, y_encoded, "BiLSTM")

# Ajout des résultats RNN au dictionnaire principal
cv_results["GRU"] = {
    "Accuracy": gru_results["Accuracy"],
    "F1-score": gru_results["F1-score"],
    "Precision": gru_results["Precision"],
    "Recall": gru_results["Recall"],
    "Training Time (s)": gru_results["Training Time"],
    "Testing Time (s)": gru_results["Testing Time"]
}

cv_results["BiLSTM"] = {
    "Accuracy": bilstm_results["Accuracy"],
    "F1-score": bilstm_results["F1-score"],
    "Precision": bilstm_results["Precision"],
    "Recall": bilstm_results["Recall"],
    "Training Time (s)": bilstm_results["Training Time"],
    "Testing Time (s)": bilstm_results["Testing Time"]
}

# === 11. Comparaison graphique des métriques ===
plt.figure(figsize=(18, 12))
scores_df = pd.DataFrame({model: {
    "Accuracy": vals["Accuracy"],
    "F1-score": vals["F1-score"],
    "Precision": vals["Precision"],
    "Recall": vals["Recall"],
    "Training Time": vals["Training Time (s)"],
    "Testing Time": vals["Testing Time (s)"]
} for model, vals in cv_results.items()}).T

# Graphique pour les scores
plt.subplot(2, 1, 1)
scores_df[["Accuracy", "F1-score", "Precision", "Recall"]].plot(kind='bar', colormap='tab10', ax=plt.gca())
plt.title("Comparaison des performances des modèles")
plt.ylabel("Score")
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Graphique pour les temps
plt.subplot(2, 1, 2)
scores_df[["Training Time", "Testing Time"]].plot(kind='bar', colormap='Pastel2', ax=plt.gca())
plt.title("Temps d'exécution des modèles")
plt.ylabel("Temps (secondes)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

# === 12. Affichage des résultats dans un tableau comparatif ===
print("\n=== Tableau comparatif des modèles ===")
results_table = pd.DataFrame(cv_results).T
results_table = results_table[['Accuracy', 'F1-score', 'Precision', 'Recall', 'Training Time (s)', 'Testing Time (s)']]
print(results_table.sort_values(by='F1-score', ascending=False))
```

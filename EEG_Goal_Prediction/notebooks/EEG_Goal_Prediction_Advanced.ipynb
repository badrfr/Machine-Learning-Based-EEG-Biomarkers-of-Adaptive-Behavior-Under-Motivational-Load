```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from timeit import default_timer as timer
from scipy import signal
from scipy.stats import entropy

from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.pipeline import make_pipeline
from sklearn.calibration import CalibratedClassifierCV
from sklearn.feature_selection import SelectKBest, f_classif, VarianceThreshold
from sklearn.impute import SimpleImputer

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GRU, Bidirectional, LSTM, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from scikeras.wrappers import KerasClassifier

# === 1. Colonnes de features et liste des fichiers ===
feature_cols = ['EEG1', 'EEG2', 'EEG3', 'EEG4', 'EEG5',
                'EEG6', 'EEG7', 'EEG8', 'Connection', 'AUX1&AUX2', 'EVENTa&b']

fichiers = [
    ('AO_10.csv', 1), ('AO_11.csv', 1), ('AO_12.csv', 1), ('AO_13.csv', 1),
    ('AO_14.csv', 1), ('AO_15.csv', 1), ('AO_16.csv', 1), ('AO_17.csv', 1),
    ('AO_18.csv', 1), ('AO_19.csv', 1), ('AO_2.csv', 1), ('AO_22.csv', 1),
    ('AO_23.csv', 1), ('AO_24.csv', 1), ('AO_3.csv', 1), ('AO_4.csv', 1),
    ('AO_6.csv', 1), ('AO_20.csv', 1), ('AO_21.csv', 1), ('AO_8.csv', 1),
    ('AO_9.csv', 1),
    ('SO_1.csv', 0), ('SO_10.csv', 0), ('SO_11.csv', 0), ('SO_12.csv', 0),
    ('SO_13.csv', 0), ('SO_14.csv', 0), ('SO_16.csv', 0), ('SO_17.csv', 0),
    ('SO_19.csv', 0), ('SO_2.csv', 0), ('SO_20.csv', 0), ('SO_23.csv', 0),
    ('SO_24.csv', 0), ('SO_25.csv', 0), ('SO_3.csv', 0), ('SO_4.csv', 0),
    ('SO_5.csv', 0), ('SO_21.csv', 0), ('SO_22.csv', 0), ('SO_6.csv', 0),
    ('SO_7.csv', 0)
]

def calculate_spectral_features(signal_data, sampling_rate=250):
    features = {}
    signal_data = signal_data.dropna()
    if len(signal_data) < 10:
        return {
            'spectral_power': 0, 'spectral_entropy': 0,
            'power_delta': 0, 'power_theta': 0, 'power_alpha': 0,
            'power_beta': 0, 'power_gamma': 0
        }
    try:
        freqs, psd = signal.welch(signal_data, fs=sampling_rate, nperseg=min(256, len(signal_data)//4))
        features['spectral_power'] = np.trapz(psd, freqs)
        psd_norm = psd / np.sum(psd)
        features['spectral_entropy'] = entropy(psd_norm)
        
        bands = {
            'delta': (0.5, 4),
            'theta': (4, 8),
            'alpha': (8, 13),
            'beta': (13, 30),
            'gamma': (30, 45)
        }
        for band, (fmin, fmax) in bands.items():
            idx = np.logical_and(freqs >= fmin, freqs <= fmax)
            band_power = np.trapz(psd[idx], freqs[idx]) if np.any(idx) else 0
            features[f'power_{band}'] = band_power
    except Exception:
        features = {
            'spectral_power': 0, 'spectral_entropy': 0,
            'power_delta': 0, 'power_theta': 0, 'power_alpha': 0,
            'power_beta': 0, 'power_gamma': 0
        }
    return features

def calculate_rms(signal_data):
    signal_data = signal_data.dropna()
    return np.sqrt(np.mean(signal_data**2)) if len(signal_data) > 0 else 0

def extraire_features(df, label=None):
    features = {}
    for col in feature_cols:
        if col in df.columns:
            series = pd.to_numeric(df[col].astype(str).str.replace(',', '.'), errors='coerce')
            if series.notna().any():
                features[f'{col}_mean'] = series.mean()
                features[f'{col}_std'] = series.std()
                features[f'{col}_min'] = series.min()
                features[f'{col}_max'] = series.max()
                features[f'{col}_median'] = series.median()
                features[f'{col}_skew'] = series.skew()
                if col.startswith('EEG'):
                    spectral = calculate_spectral_features(series)
                    features[f'{col}_spectral_power'] = spectral['spectral_power']
                    features[f'{col}_spectral_entropy'] = spectral['spectral_entropy']
                    features[f'{col}_rms'] = calculate_rms(series)
                    features[f'{col}_power_delta'] = spectral['power_delta']
                    features[f'{col}_power_theta'] = spectral['power_theta']
                    features[f'{col}_power_alpha'] = spectral['power_alpha']
                    features[f'{col}_power_beta'] = spectral['power_beta']
                    features[f'{col}_power_gamma'] = spectral['power_gamma']
            else:
                for stat in ['mean', 'std', 'min', 'max', 'median', 'skew']:
                    features[f'{col}_{stat}'] = 0
                if col.startswith('EEG'):
                    features[f'{col}_spectral_power'] = 0
                    features[f'{col}_spectral_entropy'] = 0
                    features[f'{col}_rms'] = 0
                    features[f'{col}_power_delta'] = 0
                    features[f'{col}_power_theta'] = 0
                    features[f'{col}_power_alpha'] = 0
                    features[f'{col}_power_beta'] = 0
                    features[f'{col}_power_gamma'] = 0
        else:
            for stat in ['mean', 'std', 'min', 'max', 'median', 'skew']:
                features[f'{col}_{stat}'] = 0
            if col.startswith('EEG'):
                features[f'{col}_spectral_power'] = 0
                features[f'{col}_spectral_entropy'] = 0
                features[f'{col}_rms'] = 0
                features[f'{col}_power_delta'] = 0
                features[f'{col}_power_theta'] = 0
                features[f'{col}_power_alpha'] = 0
                features[f'{col}_power_beta'] = 0
                features[f'{col}_power_gamma'] = 0
    if label is not None:
        features['Label'] = label
    return features

print("Extraction des features...")
train_data = []
for i, (fichier, label) in enumerate(fichiers):
    print(f"Traitement du fichier {i+1}/{len(fichiers)}: {fichier}")
    try:
        df = pd.read_csv(fichier, thousands='.', decimal=',', low_memory=False)
        train_data.append(extraire_features(df, label))
    except Exception as e:
        print(f"Erreur avec {fichier}: {e}")
        default_features = {f'{col}_{stat}': 0 for col in feature_cols for stat in ['mean', 'std', 'min', 'max', 'median', 'skew']}
        for col in feature_cols:
            if col.startswith('EEG'):
                default_features[f'{col}_spectral_power'] = 0
                default_features[f'{col}_spectral_entropy'] = 0
                default_features[f'{col}_rms'] = 0
                default_features[f'{col}_power_delta'] = 0
                default_features[f'{col}_power_theta'] = 0
                default_features[f'{col}_power_alpha'] = 0
                default_features[f'{col}_power_beta'] = 0
                default_features[f'{col}_power_gamma'] = 0
        default_features['Label'] = label
        train_data.append(default_features)

df_train = pd.DataFrame(train_data)
print(f"Dataset créé avec {len(df_train)} échantillons et {len(df_train.columns)-1} features")

# Préparation des données
imputer = SimpleImputer(strategy='median')
X = pd.DataFrame(imputer.fit_transform(df_train.drop('Label', axis=1)), columns=df_train.drop('Label', axis=1).columns)
y = df_train['Label']

constant_filter = VarianceThreshold()
X_filtered = pd.DataFrame(constant_filter.fit_transform(X), columns=X.columns[constant_filter.get_support()])
selector = SelectKBest(score_func=f_classif, k=min(30, X_filtered.shape[1]))
X_selected = selector.fit_transform(X_filtered, y)
selected_columns = X_filtered.columns[selector.get_support()]
X = pd.DataFrame(X_selected, columns=selected_columns)
print(f"Selected features: {list(selected_columns)}")

# Préparation pour RNN
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_rnn = X_scaled.reshape((X_scaled.shape[0], 1, X_scaled.shape[1]))
y_encoded = LabelEncoder().fit_transform(y)

# Définition des modèles
def build_gru_model(input_shape=(1, X.shape[1])):
    model = Sequential()
    model.add(GRU(64, input_shape=input_shape, dropout=0.2, recurrent_dropout=0.2))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def build_bilstm_model(input_shape=(1, X.shape[1])):
    model = Sequential()
    model.add(Bidirectional(LSTM(64, dropout=0.2, recurrent_dropout=0.2), input_shape=input_shape))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Dictionnaire des modèles
models = {
    "Random Forest": RandomForestClassifier(n_estimators=100, max_depth=5, min_samples_split=5, random_state=42),
    "XGBoost": XGBClassifier(n_estimators=100, max_depth=3, learning_rate=0.1, random_state=42, eval_metric='logloss'),
    "MLP": make_pipeline(StandardScaler(), MLPClassifier(hidden_layer_sizes=(50,), alpha=0.01, max_iter=500, random_state=42, early_stopping=True)),
    "SVM": make_pipeline(StandardScaler(), CalibratedClassifierCV(SVC(kernel='rbf', C=1, random_state=42), cv=3)),
    "Logistic Regression": make_pipeline(StandardScaler(), LogisticRegression(C=0.1, max_iter=1000, random_state=42)),
    "KNN": make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors=5, weights='distance')),
    "GRU": KerasClassifier(build_fn=build_gru_model, epochs=50, batch_size=16, verbose=0),
    "BiLSTM": KerasClassifier(build_fn=build_bilstm_model, epochs=50, batch_size=16, verbose=0)
}

# Évaluation des modèles
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_results = {}

print("\n=== Évaluation des modèles par validation croisée ===")
for name, model in models.items():
    print(f"\n--- {name} ---")
    total_train_time = 0
    total_test_time = 0
    y_pred_total = np.zeros_like(y_encoded if name in ["GRU", "BiLSTM"] else y)
    
    for train_idx, test_idx in cv.split(X_rnn if name in ["GRU", "BiLSTM"] else X, y_encoded if name in ["GRU", "BiLSTM"] else y):
        if name in ["GRU", "BiLSTM"]:
            X_train, X_test = X_rnn[train_idx], X_rnn[test_idx]
            y_train, y_test = y_encoded[train_idx], y_encoded[test_idx]
        else:
            X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
            y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

        start_train = timer()
        model.fit(X_train, y_train)
        total_train_time += timer() - start_train

        start_test = timer()
        y_pred = model.predict(X_test)
        total_test_time += timer() - start_test

        if name in ["GRU", "BiLSTM"]:
            y_pred_total[test_idx] = y_pred.flatten()
        else:
            y_pred_total[test_idx] = y_pred

    target = y_encoded if name in ["GRU", "BiLSTM"] else y
    acc = accuracy_score(target, y_pred_total)
    f1 = f1_score(target, y_pred_total)
    precision = precision_score(target, y_pred_total)
    recall = recall_score(target, y_pred_total)
    cm = confusion_matrix(target, y_pred_total)

    cv_results[name] = {
        "Accuracy": acc,
        "F1-score": f1,
        "Precision": precision,
        "Recall": recall,
        "Confusion Matrix": cm,
        "Training Time (s)": total_train_time,
        "Testing Time (s)": total_test_time
    }

    print(f"Accuracy       : {acc:.3f}")
    print(f"F1-score       : {f1:.3f}")
    print(f"Precision      : {precision:.3f}")
    print(f"Recall         : {recall:.3f}")
    print(f"Training Time  : {total_train_time:.4f} sec (total)")
    print(f"Testing Time   : {total_test_time:.4f} sec (total)")
    print("Confusion Matrix:")
    print(cm)
    print(classification_report(target, y_pred_total, target_names=["Sans objectif", "Avec objectif"]))

# Résumé final
results_table = pd.DataFrame(cv_results).T
results_table = results_table[['Accuracy', 'F1-score', 'Precision', 'Recall', 'Training Time (s)', 'Testing Time (s)']]
results_table_sorted = results_table.sort_values(by='F1-score', ascending=False)

print("\n=== Tableau comparatif ===")
print(results_table_sorted)

best_model = results_table_sorted.index[0]
print(f"\n🏆 Meilleur modèle : {best_model} (F1-score = {results_table_sorted.loc[best_model]['F1-score']:.3f})")

# === VISUALISATION DES PERFORMANCES ===
plt.figure(figsize=(18, 12))

# 1. Comparaison des métriques principales
plt.subplot(2, 2, 1)
metrics = ['Accuracy', 'F1-score', 'Precision', 'Recall']
x = np.arange(len(results_table))
width = 0.2

for i, metric in enumerate(metrics):
    plt.bar(x + i*width, results_table[metric], width, label=metric)

plt.xlabel('Modèles')
plt.ylabel('Score')
plt.title('Comparaison des performances par modèle')
plt.xticks(x + width*1.5, results_table.index, rotation=45)
plt.ylim(0, 1.1)
plt.legend()
plt.grid(axis='y', alpha=0.3)

# 2. Matrice de confusion du meilleur modèle
plt.subplot(2, 2, 2)
sns.heatmap(cv_results[best_model]["Confusion Matrix"], 
            annot=True, fmt='d', cmap='Blues',
            xticklabels=["Sans objectif", "Avec objectif"],
            yticklabels=["Sans objectif", "Avec objectif"])
plt.title(f'Matrice de confusion - {best_model}\n(Meilleur modèle)')
plt.ylabel('Vérité terrain')
plt.xlabel('Prédiction')

# 3. Comparaison des temps d'entraînement
plt.subplot(2, 2, 3)
training_times = results_table['Training Time (s)'].sort_values()
plt.barh(training_times.index, training_times, color='skyblue')
plt.xlabel('Temps (secondes)')
plt.title('Temps d\'entraînement total')
plt.grid(axis='x', alpha=0.3)

# 4. Comparaison des temps de prédiction
plt.subplot(2, 2, 4)
testing_times = results_table['Testing Time (s)'].sort_values()
plt.barh(testing_times.index, testing_times, color='lightgreen')
plt.xlabel('Temps (secondes)')
plt.title('Temps de prédiction total')
plt.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

# === CALCUL DE L'ÉNERGIE LIBRE APPROXIMATIVE ===
def compute_energy_free(model, X, y_true):
    if hasattr(model, "predict_proba"):
        try:
            probs = model.predict_proba(X)
            true_probs = probs[np.arange(len(y_true)), y_true]
            free_energy = -np.log(true_probs + 1e-8)
            return np.mean(free_energy)
        except:
            return np.nan
    return np.nan

energy_scores = {}
for name, model in models.items():
    if name not in ["GRU", "BiLSTM"]:
        try:
            model.fit(X, y)
            energy_scores[name] = compute_energy_free(model, X, y)
        except:
            energy_scores[name] = np.nan
    else:
        energy_scores[name] = np.nan  # Pas applicable pour RNN directement ici

# Ajout au tableau
results_table["Free Energy"] = results_table.index.map(energy_scores.get)
results_table_sorted = results_table.sort_values(by="F1-score", ascending=False)

# Affichage final
print("\n=== Tableau comparatif avec énergie libre ===")
print(results_table_sorted[["Accuracy", "F1-score", "Precision", "Recall", "Free Energy"]])
```
